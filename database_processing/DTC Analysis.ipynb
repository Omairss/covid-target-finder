```python
from chembl_webresource_client.new_client import new_client
molecule = new_client.molecule
```

```python
dtc_compound_list_unique_inchikeys = []
dtc_compound_list_unique_smiles = []

for chmblid in dtc_compound_list_unique:
    try:
        records = molecule.get([chmblid])
        dtc_compound_list_unique_inchikeys.append(records[0]['molecule_structures']['standard_inchi_key'])
        dtc_compound_list_unique_smiles.append(records[0]['molecule_structures']['canonical_smiles'])
    except:
        dtc_compound_list_unique_inchikeys.append(None)
        dtc_compound_list_unique_smiles.append(None)        
```

# Gene seq retreival

```python
import ensembl_rest
```

```python
import mygene

mg = mygene.MyGeneInfo()
```

```python
DTC_gene_names = list(set(dtc_df['gene_names'].tolist()))
```

```python
out = mg.querymany(DTC_gene_names, scopes='symbol', fields='entrezgene', species='human')
```

```python
len(DTC_gene_names),len(out)
```

```python
out[10]
```

```python
gene_list_DTC_dict = dict()
#entrezid_list_DTC = []
#gene_seq_list_DTC = []

for ele in out:
    #is_found = ele['notfound']
    #if is_found is True:
    #    continue
    #else:
    try:
        gene_name = ele['query']
        gene_list_DTC.append(gene_name)
        enterz_id = ele['entrezgene']
        entrezid_list_DTC.append(enterz_id)
        seq_val = ensembl_rest.sequence_id(enterz_id)
        gene_seq_list_DTC.append(seq_val)
        gene_list_DTC_dict[gene_name] = dict()
        gene_list_DTC_dict[gene_name]['enteez_id'] = enterz_id
        gene_list_DTC_dict[gene_name]['gene_seq'] = seq_val
    except:
        continue
```

```python
gene_list_DTC_dict
```

```python
f = open("DTC_gene_seq_dict.txt","w")
f.write(str(gene_list_DTC_dict))
f.close()
```

```python
len(gene_list_DTC),len(entrezid_list_DTC),len(gene_seq_list_DTC)
```

```python
seq_list = []

for index, row in dtc_df.iterrows():
    gene_id = row['gene_names']
    try:
        seq_list.append(gene_list_DTC_dict[gene_id]['gene_seq']['seq'])
    except:
        seq_list.append(None)
```

```python
dtc_df['gene_seq'] = seq_list
```

```python
dtc_df.to_pickle('data/DTC_with_gene_seq.pkl')
```
