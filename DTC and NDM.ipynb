```python
chembl_inchi_mapping = pd.read_csv('data/chembl_26_chemreps.txt', sep="\t")
```

```python
chembl_inchi_mapping
```

```python
chembl_inchi_mapping_chembl_id = chembl_inchi_mapping['chembl_id'].tolist()
chembl_inchi_mapping_inchikey = chembl_inchi_mapping['standard_inchi_key'].tolist()
```

```python
inchi_column = []

for index, row in dtc_df.iterrows():
    if row['compound_id'] in chembl_inchi_mapping_chembl_id:
        inchi_column.append(chembl_inchi_mapping_inchikey[chembl_inchi_mapping_chembl_id.index(row['compound_id'])])
    else:
        inchi_column.append(None)
```

```python
ndm_df = pd.read_pickle('NDM_Master_FingerPrints_InChi.pkl')
```

```python
ndm_inchikey_list = list(set(ndm_df['InChIKey'].tolist()))
```

```python
dtc_df_dropped = pd.read_pickle('data/DTC_with_gene_seq_dropped.pkl')
```

```python
inchi_column = []
is_in_NDM = []

for index, row in dtc_df_dropped.iterrows():
    if row['compound_id'] in chembl_inchi_mapping_chembl_id:
        Inchikeyval = chembl_inchi_mapping_inchikey[chembl_inchi_mapping_chembl_id.index(row['compound_id'])]
        inchi_column.append(Inchikeyval)
        if Inchikeyval in ndm_inchikey_list:
            is_in_NDM.append(1)
        else:
            is_in_NDM.append(0)
    else:
        inchi_column.append(None)
        is_in_NDM.append(0)    
```
